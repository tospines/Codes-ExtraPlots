```python
import os
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from astropy.io import fits
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib import rc, rcParams
from scipy import interpolate
import matplotlib.ticker as mticker
import math
import time
import scipy.optimize as optimize
from scipy.integrate import quad
```

```python
M_A = 0.3
D_ref = 3.8*1.e28 ## this is [cm^2/s]
```

```python
## Block for preliminary assignments ##
L_ref = 120.    # chosen reference length, in [pc]   Default is 120 pc!!!
print(r'Alfvenic Mach number, M_A:', M_A)

# Diffusion coefficient 
print(r'Normalization of D, D_0:', D_ref, '\n')
delta = 1./3. #0.5

Nt = int(5.67e5) 
Nz = 200
Nr = 200

Lr = 1
Lz = 1

z = np.linspace(start = -Lz, stop = Lz, num = Nz+1, endpoint=True)
r = np.linspace(start = 0, stop = Lr, num = Nr+1, endpoint=True)
#r = np.linspace(start = 0, stop = Lr, num = Nr+1, endpoint=True)

dr = Lr/Nr
dz = 2*Lz/Nz
```

```python
##############################################################################################################################

Emin = 300. #300.
Emax = 2.e5 #4.e5 
Ekin_fac = 1.2 # 1.3 
#dimE = 60
#Ekin_fac = np.exp( (np.log(Emax) - np.log(Emin)) / (dimE-1) )
dimE = int( (np.log(Emax) - np.log(Emin)) / np.log(Ekin_fac)) + 2 # I added another more point to pass E_max!
E = [Emin*(Ekin_fac**i) for i in range(0,dimE)]
E = np.array(E)
dE = (np.append(E[1:], Emin*(Ekin_fac**dimE)) - E)/E

print('Evaluating', dimE, 'Points in energy')
print('Energy grid:', list(E), 'Energy resolution:', np.unique(dE))
print('')
###############

#d_cyg_ref = 1.5e+3           # reference distance of the Cygnus-X region, in [pc]
conv_pc_cm = 3.086e+18       # conversion factor from [pc] to [cm]
conv_cm_pc = 3.24e-19        # conversion factor from [cm] to [pc]
conv_s_yr = 3.171e-8         # conversion factor from [sec] to [yr]
conv_yr_sec = 3.154e+7       # conversion factor from [yr] to [sec]
conv_rad_degree = 57.2958    # conversion factor from [rad] to [deg]
c = 2.998e+10                # speed of light, in [cm/s]



print(r'Our reference diffusion coefficient (at 1 GeV) is: D_ref =', D_ref, r'cm^2/s =', 
      D_ref*((conv_cm_pc**2)/conv_s_yr), r'$pc^2/yr$')
print("")
D = (D_ref*(E**delta)/(L_ref*c*conv_pc_cm))



# Time grid #
F = 1./2.      # according to the stability condition
dt = F / ( np.max(D)/dr**2 + np.max(D)/dz**2 )
#dt *= 20
#dt = 0.5999170828110867/(L_ref/(c*conv_cm_pc/conv_s_yr))

t = np.arange(start = 0., stop = Nt*dt, step = dt)
#############



print("Nr = ", Nr)
print("Nz = ", Nz)
print("Nt = ", Nt)
print("")

print("r array: from", np.min(r)*L_ref, 'to', np.max(r)*L_ref, 'pc')
print("z array: from", np.min(z)*L_ref, 'to', np.max(z)*L_ref, 'pc')
print("t array: from", np.min(t)*L_ref/(c*conv_cm_pc/conv_s_yr)*1e-3, 'to', np.max(t)*L_ref/(c*conv_cm_pc/conv_s_yr)*1e-3, 'kyr')
print("")

print("Space step dr =", dr*L_ref, 'pc')
print("Space step dz =", dz*L_ref, 'pc')
print("Time step dt =", dt*L_ref/(c*conv_cm_pc/conv_s_yr), 'yr')
print("")
```

```python
###########################################  Definitions for the solver #######################################################################

def tridiag(a, b, c, k1=-1, k2=0, k3=1):
    return np.diag(a, k1) + np.diag(b, k2) + np.diag(c, k3)

B_field = 3 # microGauss
k_B = 8.62e-5 # eV/K
sigmaT = 6.652e-25 # cm^2
m_ec2 = 0.511*1e6 #eV/c^2 
U_B = 0.224*((B_field/3)**2) *1e-9 # eV/cm^3 --> GeV/cm^3 and B_field in microGauss
U_CMB = 0.26 *1e-9  # eV/cm^3 --> GeV/cm^3 
U_IR = 0.25 *1e-9   # eV/cm^3 --> GeV/cm^3 
U_Opt = 0.055 *1e-9 # eV/cm^3 --> GeV/cm^3 
U_UV1 = 0.37 *1e-9  # eV/cm^3 --> GeV/cm^3 
U_UV2 = 0.23 *1e-9  # eV/cm^3 --> GeV/cm^3 
U_UV3 = 0.12 *1e-9  # eV/cm^3 --> GeV/cm^3 


def f_KN(En, RF="CMB"):
    En = En*1e6 # From GeV to eV!
    if RF == "IR":
        T = 33.07 # K
    elif RF == "Opt":
        T = 313.32 # K
    elif RF == "UV1":
        T = 3249.3 # K
    elif RF == "UV2":
        T = 6150.4 # K
    elif RF == "UV3":
        T = 23209.0 # K
    else:
        T = 2.7 # K
    return ((45/64)*(np.pi**2)*((m_ec2)/k_B*T)**2)/ ((45/64)*(np.pi**2)*((m_ec2)/k_B*T)**2 + ((En/(m_ec2*1e-9))**2))


#U_ph = U_CMB + U_Opt + U_IR + U_UV1 + U_UV2 + U_UV3 # +U_Star?
C_loss = -(4/3)*c*sigmaT/((m_ec2)**2) * (L_ref*conv_pc_cm)/c    # (L_ref/(c*conv_cm_pc/conv_s_yr))         
C_lossGeV = -(4/3)*c*sigmaT/((m_ec2*1e-9)**2) * (L_ref*conv_pc_cm)/c  
## Units of cm^3/(s GeV^2) * s (to adimensionalize time)

def b_loss(En):
    return C_lossGeV*(U_B + U_CMB*f_KN(En, "CMB") + U_Opt*f_KN(En, "Opt") +\
                + U_IR*f_KN(En, "IR") + U_UV1*f_KN(En, "UV1") + U_UV2*f_KN(En, "UV2") + U_UV3*f_KN(En, "UV3"))*(En**2)
## Unit of GeV, since C_lossGeV is multiplied by E^2 * GeV/cm^3
```

```python
## Source term
alpha = -1.6 ##2.2
Ec = 2.e5 #1e6 # GeV 
Brakn = 3
tau_0 = 1.2e4*conv_yr_sec # 12 kyr -> seconds  #1.2e6*3.1e10 # 1.20 Myr -> seconds    #1.2e4*3.1e10 # 12 kyr -> seconds
Edot = 3.2e34 # erg/s  # Geminga-like. Taken from various references, among them arXiv:2107.07395v1 
L_0 = Edot * 624.151# L/eta at t = 0 -- erg/s -> GeV/s
Q_0 = 3.2 # d^{2}N/dE^{2} -- particles/GeV^2

C_e = 2.335e-2 * 4*np.pi/(c*10000)# 1/(m^2 sr s GeV)  -> 1/(cm^3 GeV)
gamm_e = -3.28
DeltaGamm_e = 0.094
Bound_flux = C_e*((E/20.04)**gamm_e) * ((E/20.04)**DeltaGamm_e)  # From PRL 122, 101101 (2019)
```

```python
############# Definition of the density grid ############
U_matrix = np.zeros((len(r), len(z), dimE+2), dtype=float)
# density is going to be computed in a square from the 2nd bin to the max-1 bin (from i=1 to i=-2=dimE)

#U_matrix[0, :, 1:-1] += Bound_flux  # Lower R
#U_matrix[-1, :, 1:-1] += Bound_flux  # Upper R
#U_matrix[:, 0, 1:-1] += Bound_flux   # Lower Z
#U_matrix[:, -1, 1:-1] += Bound_flux  # Upper Z

r_cent = np.where(r==0)[0][0]
z_cent = np.where(z==0)[0][0]    
def Source(tt):    #dN/dtdE from L(dE/dt)*Q_0(dN/dE^2)
    Source_map = np.zeros(U_matrix.shape) 
    Source_map[r_cent, z_cent, 1:dimE+1] +=\
    (L_ref*conv_pc_cm/c)*L_0*(1+(tt*L_ref*conv_pc_cm/c/tau_0))**(-(Brakn+1)/(Brakn-1)) * Q_0*((E)**(alpha))*np.exp(-E/Ec)  
    #(i.e. a row in the density matrix will consist of 0,vec(x),0, this is why it has len of Nx+2)
    #(L_ref/(c*conv_cm_pc)) is the adimensionalization factor
    return Source_map
```

```python
################# Crank-Nicholson coefficients ####################
# terms for first equation (t-> t+Dt/2)
A_CN1 = 1 + 2*D*dt/(2*dz**2)
B_CN1 = -1*D*dt/(2*dz**2)
C_CN1 = (1 - 2*(M_A**4)*D*dt/(2*dr**2)) ## It is divided by two because we are using dt/2!!!
    
G_CN1 = np.zeros(len(E)+1) #-1*b_loss(E)*dt/(np.append(E[1:], Emin*(Ekin_fac**dimE)) - E)
         
            
E_CN1 = []
D_CN1 = []
for ie in range(dimE):                
    MyE_CN1 = (M_A**4)*D[ie]*dt/(2*dr**2) - (D[ie]*(M_A**4)*dt/(2*2*dr*(r+dr/2)))
    #MyE_CN1[r_cent] = 0. #(M_A**4)*D[ie]*dt/(2*dr**2)/2 - (D[ie]*(M_A**4)*dt/(2*2*dr*dr))/2
    E_CN1.append(MyE_CN1)
    MyD_CN1 = (M_A**4)*D[ie]*dt/(2*dr**2) + (D[ie]*(M_A**4)*dt/(2*2*dr*(r+dr/2)))
    #MyD_CN1[r_cent] = 2*(M_A**4)*D[ie]*dt/(2*dr**2) #(M_A**4)*D[ie]*dt/(2*dr**2) + (D[ie]*(M_A**4)*dt/(2*2*dr*dr))
    D_CN1.append(MyD_CN1)
    # Create a matrix, for elementwise multiplication! 
    #it has shape (dimE, Nr, Nz) and we want to multiply by (Nr, Nz, dimE), so B_2D.T[:, :, 1] should be (Nx, Ny, dimE)

E_CN1 = np.array(E_CN1)
D_CN1 = np.array(D_CN1)



# terms for second equation (t+Dt/2 -> t+Dt)
A_CN2 = 1 + 2*(M_A**4)*D*dt/(2*dr**2)
D_CN2 = (1 - 2*D*dt/(2*dz**2) )
E_CN2 = D*dt/(2*dz**2)
#G_CN2 = b_loss(E)*dt/(2*(np.append(E[1:], Emin*(Ekin_fac**dimE)) - E))/2 ## Not necessary!!!!

C_CN2 = []
B_CN2 = []
for ie in range(dimE):
    MyC_CN = -1*(M_A**4)*D[ie]*(dt/(2*dr**2) - (dt/(2*2*dr*(r+dr/2))) )
    #MyC_CN[r_cent] = 0 #-1*(M_A**4)*D[ie]*(dt/(2*dr**2) - (dt/(2*2*dr*dr)) )
    C_CN2.append(MyC_CN)
    MyB_CN = -1*(M_A**4)*D[ie]*(dt/(2*dr**2) + (dt/(2*2*dr*(r+dr/2))) )
    #MyB_CN[r_cent] = -2*(M_A**4)*D[ie]*dt/(2*dr**2)#-1*(M_A**4)*D[ie]*(dt/(2*dr**2) + (dt/(2*2*dr*dr)) )
    B_CN2.append(MyB_CN)
    # Create a matrix, for elementwise multiplication! 
    #it has shape (dimE, Ny, Nx) and we want to multiply by (Nx, Ny, dimE), so B_2D.T[:, :, 1] should be (Nx, Ny, dimE)

C_CN2 = np.array(C_CN2)
B_CN2 = np.array(B_CN2)

print(r_cent, z_cent)
```

## Main method

```python
#### Running!!! #########
print('Running the solver!\n')

t0 = time.time() 

simsteps = int(1e3) #Nt
S_ = []
Inject = []
print('Total simulation time:', np.round(simsteps*dt *L_ref*conv_pc_cm/c *conv_s_yr /1000, 2), 'kyrs')
for it in range(simsteps): 
    for ie in range(dimE, 0, -1):
        
        Alpha_CN1 = tridiag(np.ones(Nz)*B_CN1[ie-1], np.ones(Nz+1)*A_CN1[ie-1], np.ones(Nz)*B_CN1[ie-1])
        Beta_CN1 = tridiag(E_CN1[ie-1][1:], np.ones(Nr+1)*C_CN1[ie-1], D_CN1[ie-1][:-1])
        
        Alpha_CN2 = tridiag(C_CN2[ie-1][1:], np.ones(Nr+1)*A_CN2[ie-1], B_CN2[ie-1][:-1])
        Beta_CN2 = tridiag(np.ones(Nz)*E_CN2[ie-1], np.ones(Nz+1)*D_CN2[ie-1], np.ones(Nz)*E_CN2[ie-1])
        
        Alpha_CN2[0][1] = -2*(M_A**4)*D[ie-1]*(dt/(2*dr**2)) ## Von Neumann boundary conditions (insulated b.c.) at r=0, 
        Beta_CN1[0][1] = 2*(M_A**4)*D[ie-1]*(dt/(2*dr**2))
        
        if it == 0:
            S_av = dt/2*(Source(t[it])[:, :, ie] + Source((t[it] + dt/2))[:, :, ie])/2
            S_av2 = dt/2*(Source((t[it] + dt))[:, :, ie] + Source((t[it] + dt/2))[:, :, ie])/2
            Inject.append(np.sum(S_av) + np.sum(S_av2))

        else: S_av = 0; S_av2 = 0
        
        ##First equation:
        U_matrix[:, :, ie] = np.dot( np.dot(Beta_CN1, U_matrix[:, :, ie]) + S_av, np.linalg.inv(Alpha_CN1.T))

        ##Second equation:
        U_matrix[:, :, ie] = np.dot(np.linalg.inv(Alpha_CN2), np.dot(Beta_CN2, U_matrix[:, :, ie].T).T + S_av2)
        
    S_.append(U_matrix.copy())

print((time.time() - t0)/3600, 'h')           
```
